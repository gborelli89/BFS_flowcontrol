# Data structure (with examples)

This notebook presents the data structure related to the cylinder position relative to a step (origin). This is part of a PhD study on passive drag control using circular cylinders. All the algorithms were developed using Julia language. 

One, two and three cylinders were considered for the studies. 

```julia
using Images
```

```julia
# Incluindo script
include("../src/structure.jl")
```

## One cylinder

For one control cylinder, the structure created is given by the cartesian coordinates. The origin is user defined. On this study, the step edge was used as origin (red dot of the figure below).

```julia
display("image/png", read("figuras/pos1cil.png"))
```

The type `OneCylinderType` was created. The parameters are the coordinates $x$ and $y$ (*Float64* type) shown above.

```julia
c1 = OneCylType(1.2,0.0)
```

## Two cylinders

When considering thw control cylinders, one can adopt different strategies. One possible strategy is to work with absolute coordinates for both cylinders (the step edge as global origin). A *Genetic Algorithm* was proposed on this research (in order to reduce search total time during the experiments). It was observed that using absolute coordinates for both cylinders could complicate the algorithm implementation. 

Therefore, it was decided to work with absolute coordinates just for the cylinder closer to the step edge. For the other cylinders relative coordinates were adopted as represented in the figure below.

```julia
display("image/png", read("figuras/pos2cil.png"))
```

A `TwoCylinderType` was created. This time the paramenters given are: $x$, $y$, $dx$ and $dy$, where $x$ and $y$ are the first downstream cylinder coordinates relative to the step edge, $dx$ and $dy$ are the second cylinder coordinates relative to the first one. All the coordinater given are *Float64*.

```julia
c2 = TwoCylType(1.2,0.0,5.2,2.0)
```

## Three cylinders

The last situation explored was the case of three control cylinders. The strategy adopted was analogous to the two control cylinders.

```julia
display("image/png", read("figuras/pos3cil.png"))
```

A  `ThreeCylType` was created with the given parameters: $x$, $y$, $dx1$, $dy1$, $dx2$ and $dy2$, all *Float64*.

```julia
c3 = ThreeCylType(1.2,0.0,5.2,2.0,2.4,1.0)
```
